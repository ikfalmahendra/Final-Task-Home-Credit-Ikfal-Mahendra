```python
# Impor library dasar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Atur agar pandas menampilkan semua kolom (opsional)
pd.set_option('display.max_columns', None)

# Muat file data latih (training data)
# KARENA KITA SUDAH UPLOAD, path-nya HANYA NAMA FILE
try:
    df_train = pd.read_csv('application_train.csv')
    df_test = pd.read_csv('application_test.csv')
    print("Data berhasil dimuat.")
except FileNotFoundError:
    print("Error: File tidak ditemukan. Pastikan Anda sudah mengunggah file .csv")
```

```python
# Tampilkan 5 baris pertama
print("Tampilan 5 baris pertama (head):")
print(df_train.head())

# Lihat informasi umum: jumlah baris, kolom, dan tipe data
print("\nInformasi Data (info):")
df_train.info()
```

```python
# Visualisasikan Target
print("\nDistribusi Target (Persentase):")
print(df_train['TARGET'].value_counts(normalize=True) * 100)

sns.countplot(x='TARGET', data=df_train)
plt.title('Distribusi Target (0 = Lunas, 1 = Gagal Bayar)')
plt.show()
```

```python
# Hitung persentase nilai kosong untuk setiap kolom
missing_data = df_train.isnull().sum().sort_values(ascending=False)
missing_percentage = (missing_data / len(df_train)) * 100

print("Persentase Nilai Kosong (20 teratas):")
print(missing_percentage.head(20))

# 1. DROP (Buang) Kolom
cols_to_drop = missing_percentage[missing_percentage > 50].index
# Kita perlu menyimpan ID Klien untuk data test nanti
test_ids = df_test['SK_ID_CURR']

df_train = df_train.drop(columns=cols_to_drop)
# Pastikan df_test juga membuang kolom yang sama
df_test = df_test.drop(columns=cols_to_drop)
print(f"\nMembuang {len(cols_to_drop)} kolom karena >50% data hilang.")

# 2. IMPUTE (Isi) Kolom
# Isi 'AMT_ANNUITY' (numerik) dengan nilai tengah (median)
median_annuity = df_train['AMT_ANNUITY'].median()
df_train['AMT_ANNUITY'].fillna(median_annuity, inplace=True)
df_test['AMT_ANNUITY'].fillna(median_annuity, inplace=True)

# Isi 'OCCUPATION_TYPE' (kategorikal) dengan 'Unknown'
df_train['OCCUPATION_TYPE'].fillna('Unknown', inplace=True)
df_test['OCCUPATION_TYPE'].fillna('Unknown', inplace=True)

# Isi 'EXT_SOURCE_2' dan 'EXT_SOURCE_3' (penting!) dengan median
df_train['EXT_SOURCE_2'].fillna(df_train['EXT_SOURCE_2'].median(), inplace=True)
df_test['EXT_SOURCE_2'].fillna(df_train['EXT_SOURCE_2'].median(), inplace=True)

df_train['EXT_SOURCE_3'].fillna(df_train['EXT_SOURCE_3'].median(), inplace=True)
df_test['EXT_SOURCE_3'].fillna(df_train['EXT_SOURCE_3'].median(), inplace=True)

print("Selesai menangani nilai kosong.")
```

```python
# --- LANGKAH 1 (Lanjutan): BLOK 5 & 6 ---

# 5. Menangani Anomali DAYS_EMPLOYED
print("Memperbaiki anomali DAYS_EMPLOYED...")
# Ganti nilai 365243 (1000 tahun) dengan 'Not a Number' (NaN) agar dianggap kosong
df_train['DAYS_EMPLOYED'].replace(365243, np.nan, inplace=True)
df_test['DAYS_EMPLOYED'].replace(365243, np.nan, inplace=True)

# Sekarang, isi nilai kosong yang baru saja kita buat (termasuk yang asli) dengan median
median_days_employed = df_train['DAYS_EMPLOYED'].median()
df_train['DAYS_EMPLOYED'].fillna(median_days_employed, inplace=True)
df_test['DAYS_EMPLOYED'].fillna(median_days_employed, inplace=True)
print("Selesai menangani anomali DAYS_EMPLOYED.")


# 6. Encoding Variabel Kategori (Mengubah Teks jadi Angka)
print("\nMemulai Encoding Kategori...")

# Dapatkan daftar semua kolom 'object' (teks) yang tersisa
categorical_cols = df_train.select_dtypes('object').columns
print(f"Kolom Kategorikal yang akan di-encode: {list(categorical_cols)}")

# Gunakan pd.get_dummies untuk One-Hot Encoding semua kolom object sekaligus
# Ini akan otomatis membuat kolom baru (mis: CODE_GENDER_M, CODE_GENDER_F)
df_train = pd.get_dummies(df_train, columns=categorical_cols)
df_test = pd.get_dummies(df_test, columns=categorical_cols)

# --- Menyamakan Kolom Train dan Test ---
# Ini langkah KRUSIAL. Kadang data test punya kategori yg tidak ada di train (atau sebaliknya)
# Kita harus pastikan kedua tabel punya kolom yang PERSIS SAMA.

print("\nMenyamakan kolom data train dan test...")
# 1. Simpan label Target (karena akan hilang saat 'align')
train_labels = df_train['TARGET']

# 2. 'align' akan menyamakan kolom.
# join='inner' berarti HANYA ambil kolom yang ada di KEDUA tabel.
df_train, df_test = df_train.align(df_test, join='inner', axis=1)

# 3. Kembalikan label Target ke df_train
df_train['TARGET'] = train_labels

print(f"\nBentuk data train baru (setelah encoding): {df_train.shape}")
print(f"Bentuk data test baru (setelah encoding): {df_test.shape}")
print("\n--- LANGKAH 1 SELESAI ---")
print("Data 'application_train' dan 'application_test' sudah bersih dan semua numerik.")
```

```python
# Lihat informasi umum: jumlah baris, kolom, dan tipe data
print("\nInformasi Data (info):")
df_train.info()
```

```python
# 1. Memisahkan TARGET dari data latih (df_train)
# X_train akan berisi semua 192 kolom fitur
# y_train hanya akan berisi 1 kolom target
X_train = df_train.drop('TARGET', axis=1)
y_train = df_train['TARGET']

# 2. Menetapkan data tes
# Kita asumsikan df_test adalah data tes yang sudah di-encode
X_test = df_test

# 3. SANGAT PENTING: Menyamakan urutan kolom
# Terkadang setelah encoding, urutan kolom X_train dan X_test berbeda.
# Kode ini memastikan keduanya punya kolom yang sama persis dalam urutan yang sama.
# Ini akan menghapus kolom 'TARGET' dari X_train (yang sudah kita pisahkan)
# dan memastikan X_test cocok.
X_train, X_test = X_train.align(X_test, join='inner', axis=1)

# 4. Tampilkan hasil untuk konfirmasi
print(f"Data siap untuk di-scale!")
print(f"Bentuk X_train (fitur latih): {X_train.shape}")
print(f"Bentuk y_train (target latih): {y_train.shape}")
print(f"Bentuk X_test (fitur tes):    {X_test.shape}")
```

```python
# --- PERBAIKAN TAHAP 2.1 ---

# 1. BUANG KOLOM ID DARI DATA LATIH DAN DATA TES
# Kita harus membuangnya dari KEDUA set data.
if 'SK_ID_CURR' in X_train.columns:
    X_train = X_train.drop('SK_ID_CURR', axis=1)
    print("Kolom SK_ID_CURR telah dibuang dari X_train.")

if 'SK_ID_CURR' in X_test.columns:
    X_test = X_test.drop('SK_ID_CURR', axis=1)
    print("Kolom SK_ID_CURR telah dibuang dari X_test.")


# 2. Logika BARU: Identifikasi kolom yang akan di-scale
# Kita asumsikan kolom yang perlu di-scale adalah kolom yang
# memiliki LEBIH DARI 2 nilai unik (bukan biner 0/1)
cols_to_scale = [col for col in X_train.columns if X_train[col].nunique() > 2]

# 3. Dapatkan daftar kolom yang TIDAK di-scale (kolom biner/flags)
cols_not_scaled = [col for col in X_train.columns if X_train[col].nunique() <= 2]

# 4. Tampilkan hasilnya untuk diperiksa
print("--------------------------------------------------")
print(f"Total kolom setelah buang ID: {len(X_train.columns)}")
print(f"Jumlah kolom yang AKAN di-scale (nunique > 2): {len(cols_to_scale)}")
print(f"Jumlah kolom yang TIDAK di-scale (binary 0/1): {len(cols_not_scaled)}")
print("--------------------------------------------------")

print("\nkolom yang AKAN di-scale:")
print(cols_to_scale)

print("\nkolom yang TIDAK di-scale:")
print(cols_not_scaled)
```

```python
# 1. Impor library-nya (jika Anda belum melakukannya di atas)
from sklearn.preprocessing import StandardScaler

# 2. Buat objek (alat) scaler-nya
#    Kita beri nama variabel ini 'scaler'
scaler = StandardScaler()

# 3. "Latih" scaler (menggunakan .fit())
#    Perintah ini memberitahu 'scaler' untuk:
#    "Hei, lihat data X_train, tapi HANYA kolom-kolom di 'cols_to_scale'.
#     Tolong hitung dan SIMPAN nilai rata-rata (mean) dan
#     standar deviasi (std) dari setiap kolom itu."
print("Memulai proses 'fit' (latih) scaler...")
print(f"Scaler akan mempelajari statistik dari {len(cols_to_scale)} kolom...")

# Ini adalah perintah inti dari Tahap 2.2:
scaler.fit(X_train[cols_to_scale])

print("-----------------------------------")
print(" BERHASIL!")
print("Scaler telah berhasil dilatih (di-'fit').")
print("Objek 'scaler' sekarang sudah 'pintar' dan siap digunakan untuk 'transform'.")
```

```python
# 1. Ubah (transform) kolom-kolom di X_train
#    Kode ini mengambil 34 kolom dari X_train, mengubah nilainya
#    menggunakan 'scaler', lalu memasukkan kembali nilai baru itu
#    ke X_train.
print("Menerapkan scaling (transform) ke X_train...")
X_train[cols_to_scale] = scaler.transform(X_train[cols_to_scale])

# 2. Ubah (transform) kolom-kolom di X_test
#    PENTING: Kita menggunakan 'scaler' yang SAMA (yang dilatih
#    pakai X_train) untuk mengubah X_test.
#    Ini adalah aturan "tidak boleh menyontek".
print("Menerapkan scaling (transform) ke X_test...")
X_test[cols_to_scale] = scaler.transform(X_test[cols_to_scale])

print("-----------------------------------")
print(" BERHASIL!")
print("Scaling (transform) selesai untuk kedua set data.")
print("DataFrames X_train dan X_test Anda kini telah diperbarui.")
```
